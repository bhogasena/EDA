# Calculate Literay Rates

```python
census_df['literacy_rate']=(census_df['Total_literate']/census_df['Total_Population'])*100
```

```python
census_df.head()
```

```python
fig= px.box(census_df,x='Age-group',y='literacy_rate')
fig.show()
```

```python
census_df['Female_Literacy_Rate'] = (census_df['Female_literate']/census_df['Total_Population'])*100
```

```python
fig=px.box(census_df,x='Area Name',y='Female_Literacy_Rate')
fig.show()
```

**Kerala is at the top of the list of female literacy rates across the states\**

```python
fig=px.bar(data_frame=census_df.groupby(['Area Name']).mean().reset_index(),x='Area Name',y='Female_Literacy_Rate')
fig.show()
```

**Bihar shows the lowest literacy rate among all the states**
