**Matplotlib**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%reload_ext autoreload
%autoreload 2 
```

**Functional Plot**

```python
x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y)
plt.title('Squares')
plt.xlabel('Numbers')
plt.ylabel('Square of Numbers')
```

**Multiple Plots**

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.xlabel('1x')


plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.xlabel('2x')

plt.subplot(2,2,3)
plt.plot(x,y,'r')
plt.xlabel('3x')

plt.subplot(2,2,4)
plt.plot(x,y,'g')
plt.xlabel('4x')
```

**Figure Objects**

```python
fig = plt.figure(figsize=(5,5),dpi=100)
axes_1 = fig.add_axes([0.1,0.1,0.5,0.5])
axes_1.set_xlabel('Numbers')
axes_1.set_ylabel('Squares of Numbers')
axes_1.set_title('Squares Plot')
axes_1.plot(x,y,label='x/x2')
axes_1.plot(y,x,label='x2/x')
axes_1.legend(loc=0)

axes_2 = fig.add_axes([0.8,0.1,0.7,0.5])
axes_2.set_xlabel('Numbers')
axes_2.set_ylabel('Squares of Numbers')
axes_2.set_title('Squares Plot')
axes_2.plot(x,y,'r',label='x/x2')
axes_2.legend(loc=0)
axes_2.set_title('Test Plot')

```

**Sub Plots**

```python
fig, axes = plt.subplots(figsize=(12,4),nrows=1,ncols=3)
plt.tight_layout()
def subplot(index,title):
  axes[index].set_title(title)
  axes[index].set_xlabel('Numbers')
  axes[index].set_ylabel('Squares of Numbers')
  axes[index].plot(x,y)

for i in range(0,3):
  subplot(i,'Plot'+(str(i)))

```
